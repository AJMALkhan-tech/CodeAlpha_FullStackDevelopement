```python
def calculate_grade(marks):
    total = sum(marks)
    percentage = total / (len(marks))

    if percentage >= 90:
        grade = "A+"
    elif percentage >= 80:
        grade = "A"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    elif percentage >= 50:
        grade = "D"
    else:
        grade = "F"

    return {
        "Total Marks": total,
        "Percentage": percentage,
        "Grade": grade
    }

marks = [79, 90, 71, 68, 98]
result = calculate_grade(marks)

print("Result Summary:")
for k, v in result.items():
    print(f"{k}: {v}")
```

## **Breadth-First Search (BFS)**

### **Introduction**

BFS is a graph traversal algorithm.

It starts from a source node and explores all neighbors before moving to the next level.

BFS is implemented using a Queue (FIFO).

### **BFS Algorithm (Steps)**

Start with a queue and enqueue the starting node.

Mark the starting node as visited.

While the queue is not empty:

    Dequeue a node.

    Visit all its unvisited neighbors and enqueue them.

Continue until all nodes are visited.

## **BFS in Python**

###**CODES I PRACTICED IN LAB (FOR BFS)**

```python
def bfs(my_graph, begin):
    seen = []
    line = [begin]

    while line:
        current = line.pop(0)
        if current not in seen:
            print(current, end=" ")
            seen.append(current)
            for nxt in my_graph[current]:
                if nxt not in seen:
                    line.append(nxt)

my_graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("BFS Traversal:")
bfs(my_graph, 'A')
```

## **Task 1: Write a function to calculate the factorial of a number.**

```python
def fact(n):
    r = 1
    for i in range(1, n+1):
        r *= i
    return r

print("Factorial of 4 is:", fact(4))
```

## **Task 2: Write a function that takes a list as input and returns the maximum element.**

```python
def findmax(values):
    biggest = values[0]
    for item in values:
        if item > biggest:
            biggest = item
    return biggest

nums = [12, 45, 7, 89, 23]
print("Maximum number is:", findmax(nums))
```

## **Task 3: Write a function that checks if a number is prime or not.**

```python
def check_prime(number):
    if number <= 1:
        return False
    for divider in range(2, number):
        if number % divider == 0:
            return False
    return True

num = 17
print("BRO is", num, "a prime number? : ", check_prime(num))
```

## **Task 1: Implement BFS on a graph given by the user.**

```python
def bfs_walk(network, source):
    seen = []
    line = [source]

    while line:
        current = line.pop(0)
        if current not in seen:
            print(current, end=" ")
            seen.append(current)
            for link in network[current]:
                if link not in seen:
                    line.append(link)

tree = {
    'X': ['Y', 'Z'],
    'Y': ['P', 'Q'],
    'Z': ['R'],
    'P': [],
    'Q': ['R'],
    'R': []
}

print("BFS graph(i named tree lol ) is starting from X:")
bfs_walk(tree, 'X')
```

## **Task 2: Modify BFS to count the number of nodes in each level.**

```python
def bfs_levels(my_map, begin):
    done = []
    line = [begin]
    step = 0

    while line:
        size = len(line)
        print("Step", step, ":", end=" ")

        for i in range(size):
            spot = line.pop(0)
            if spot not in done:
                print(spot, end=" ")
                done.append(spot)
                for buddy in my_map[spot]:
                    if buddy not in done:
                        line.append(buddy)
        print()
        step += 1


tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("BFS by level count is :")
bfs_levels(tree, 'A')
```

## **Task 3: Write a BFS function that checks if there is a path between two given nodes.**

```python
def path_exists(my_map, begin, target):
    seen = []
    line = [begin]

    while line:
        spot = line.pop(0)
        if spot == target:
            return True
        if spot not in seen:
            seen.append(spot)
            for buddy in my_map[spot]:
                if buddy not in seen:
                    line.append(buddy)
    return False


tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("Checking  for  A to F:", path_exists(tree, 'A', 'F'))
print("Checking for D to C:", path_exists(tree, 'D', 'C'))
```
